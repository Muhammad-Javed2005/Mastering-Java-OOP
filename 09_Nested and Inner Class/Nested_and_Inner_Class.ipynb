# **Nested and Inner Classes**

---

## **1. Introduction to Nested Classes**

- Definition: A class declared inside another class.
- Purpose: Logical grouping, better encapsulation, readable code.
- Types of Nested Classes:
- - **Static Nested Class**
- - **Non-static Nested Class (Inner Class)**

---


## **2. Static Nested Class**

- Declared as static.
- Can access only static members of the outer class.
- Does not require an object of outer class.
- Example use case: Helper/utility classes

---


## **3. Inner Classes (Non-static Nested Classes)**

- Inner classes are directly associated with an object of outer class. - Types:

### **3.1 Member Inner Class**

- Defined inside a class (non-static).
- Can access all members (including private) of outer class.
- Needs outer class object to create

### **3.2 Local Inner Class**

- Declared inside a method, constructor, or block.
- Scope is limited to that block.
- Useful for helper tasks inside a method

### **3.3 Anonymous Inner Class**

- Class without a name, created at runtime.
- Used to implement an interface or override methods of a class quickly.
- Very common in event handling (e.g., GUI, listeners).


---


## **4. Access Rules of Inner Classes**

- Inner class can access private fields of outer class.
- Outer class needs object reference to access inner class.
- Static vs Non-static rules difference


---


## **5. Advantages of Nested and Inner Classes**

- Improves encapsulation.
- Makes code more readable and maintainable.
- Used in event-driven programming.
- Helps in grouping logically related classes.


---

---


## **7. Best Practices**

- Use inner classes only when they are tightly bound to outer class.
- Prefer static nested class if inner class does not need access to outer class instance.
- Use anonymous inner class for single-use implementations.
